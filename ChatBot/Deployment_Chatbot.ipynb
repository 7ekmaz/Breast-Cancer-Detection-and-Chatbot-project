```python
%%writefile app.py
# Save as app.py
import re
import numpy as np
from PIL import Image
import tensorflow as tf
import streamlit as st
from PyPDF2 import PdfReader
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from transformers import pipeline
from deep_translator import GoogleTranslator
from streamlit_extras.add_vertical_space import add_vertical_space
import logging
from tensorflow.keras.models import load_model

# Set up logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger()

# Define the background image using HTML
background_img_html = '''
    <style>
    .stApp {
        background-image: url("https://i.imgur.com/hyzsSeC.jpeg");
        background-size: cover;
        background-position: center;
        background-repeat: no-repeat;
    }
    </style>
'''

# Inject the custom background image style
st.markdown(background_img_html, unsafe_allow_html=True)

# Streamlit app
st.title("ChatBot Demo")

# --- Q&A Section ---
st.subheader("Ask a Question")

# Upload a PDF file
pdf = '/content/drive/MyDrive/NeuroTech Grad_Project/Breast Cancer QnA.pdf'

# Function to extract all text from a PDF
def extract_text_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text()
    return text

# Function to extract questions and answers using regex
def extract_questions_and_answers(text):
    questions = re.findall(r"(Q\d+:\s*(.*?))\s*A\d+:", text, re.DOTALL)
    answers = re.findall(r"A\d+:\s*(.*?)(?=Q\d+:|$)", text, re.DOTALL)
    qa_pairs = [{"question": q.strip(), "answer": a.strip()} for (_, q), a in zip(questions, answers)]
    return qa_pairs

# Function to load the PDF, extract text and questions/answers
def load_pdf_qa_pairs(pdf_path):
    text = extract_text_from_pdf(pdf_path)
    qa_pairs = extract_questions_and_answers(text)
    return qa_pairs

# Load a domain-specific SentenceTransformer model
embedding_model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')

# Function to extract question embeddings
def generate_question_embeddings(qa_pairs):
    questions = [pair['question'] for pair in qa_pairs]
    question_embeddings = embedding_model.encode(questions)
    return question_embeddings, questions

# Function to find similar questions with a similarity threshold
def find_similar_questions(user_question, question_embeddings, questions, top_k=1, similarity_threshold=0.5):
    user_embedding = embedding_model.encode([user_question])
    similarities = cosine_similarity(user_embedding, question_embeddings)[0]
    sorted_indices = np.argsort(similarities)[::-1]
    top_indices = [i for i in sorted_indices if similarities[i] > similarity_threshold][:top_k]
    return [(questions[i], similarities[i], i) for i in top_indices]

# Function to generate a broader combined context
def generate_combined_context(user_question, qa_pairs, question_embeddings, questions, top_k=1):
    similar_questions = find_similar_questions(user_question, question_embeddings, questions, top_k)
    if similar_questions:
        related_context = " ".join([qa_pairs[i]['answer'] for _, _, i in similar_questions])
    else:
        related_context = "No relevant context found in the corpus."
    return related_context

# Load the Q&A pipeline with an advanced model
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")

# Function to get a related answer using Hugging Face Q&A pipeline
def get_related_answer(user_question, qa_pairs, question_embeddings, questions):
    if user_question.lower() == "yes":
        # Raise a flag or log the query
        logger.info("The user entered 'yes'. Raising a flag for developers.")
        st.warning("Flag raised: User entered 'yes'.")
        return "Thank you for your confirmation."

    related_context = generate_combined_context(user_question, qa_pairs, question_embeddings, questions)
    if related_context == "No relevant context found in the corpus.":
        return related_context

    result = qa_pipeline(
        question=user_question,
        context=related_context,
        max_answer_len=100,  # Allow longer answers
        handle_impossible_answer=True  # Handle cases with no valid answer
    )
    if result['score'] < 0.99:  # Low confidence score fallback
        return f" {related_context[:900]}"
    return result['answer']

if pdf is not None:
    # Load QA pairs from the PDF
    qa_pairs = load_pdf_qa_pairs(pdf)

    # Generate question embeddings
    question_embeddings, questions = generate_question_embeddings(qa_pairs)

    # Ask the user to input their question
    user_question = st.text_input("What would you like to know about?")
    if user_question:
        # Get the related answer from the model
        answer = get_related_answer(user_question, qa_pairs, question_embeddings, questions)
        # Display the question and answer
        st.write("Question:", user_question)
        st.write("Answer:", answer)

        # Language selection form for translation
        with st.form("language_selection"):
            st.write("Select language for the answer:")
            dest_lang = st.selectbox("Select destination language:", ["en", "fr", "de", "es", "it", "ar"])
            submit_button = st.form_submit_button("Translate and Display Answer")

        # Handle translation
        if submit_button:
            try:
                translation = GoogleTranslator(source='auto', target=dest_lang).translate(answer)
                st.write(f"Translated Answer ({dest_lang}): {translation}")
            except Exception as e:
                st.error(f"An error occurred during translation: {str(e)}")
```

```python
# Save as app.py
%%writefile app.py

import re
import numpy as np
from PyPDF2 import PdfReader
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from transformers import pipeline
from deep_translator import GoogleTranslator
from streamlit_extras.add_vertical_space import add_vertical_space
import streamlit as st
import logging

# --- Logging Setup ---
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger()

# --- Load AI Models ---
@st.cache_resource
def load_models():
    try:
        logger.info("Loading embedding model and QA pipeline...")

        # Load the Sentence Transformer for embedding
        embedding_model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
        logger.info("Embedding model loaded.")

        # Load the QA pipeline
        qa_pipeline = pipeline("question-answering", model="deepset/deberta-v3-base-squad2", tokenizer="deepset/deberta-v3-base-squad2")
        logger.info("QA pipeline loaded.")

        return embedding_model, qa_pipeline
    except Exception as e:
        logger.error(f"Error loading models: {e}")
        st.error("An error occurred while loading the models. Please check the logs.")
        return None, None

embedding_model, qa_pipeline = load_models()

# --- Custom Styles ---
background_img_html = '''
    <style>
    .stApp {
        background-image: url("https://i.imgur.com/hyzsSeC.jpeg");
        background-size: cover;
        background-position: center;
        background-repeat: no-repeat;
    }
    .stButton>button {
        background-color: #4CAF50;
        color: white;
        padding: 20px 24px;
        border-radius: 8px;
        font-size: 18px;
        font-weight: bold;
        box-shadow: 2px 2px 10px rgba(0, 0, 0, 0.2);
        transition: background-color 0.3s ease;
    }
    .stButton>button:hover {
        background-color: #45a049;
    }
    </style>
'''
st.markdown(background_img_html, unsafe_allow_html=True)

# --- Question Answering Section ---
st.title("PDF-Based Chatbot")

# PDF file containing FAQs
pdf_path = '/content/drive/MyDrive/ChatBot LoadStation/LoadStation fq.pdf'

def extract_text_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    text = "".join(page.extract_text() for page in reader.pages)
    return text

def extract_questions_and_answers(text):
    questions = re.findall(r"(Q\d+:\s*(.*?))\s*A\d+:", text, re.DOTALL)
    answers = re.findall(r"A\d+:\s*(.*?)(?=Q\d+:|$)", text, re.DOTALL)
    qa_pairs = [{"question": q.strip(), "answer": a.strip()} for (_, q), a in zip(questions, answers)]
    return qa_pairs

@st.cache_data
def load_qa_pairs(pdf_path):
    text = extract_text_from_pdf(pdf_path)
    return extract_questions_and_answers(text)

if pdf_path:
    qa_pairs = load_qa_pairs(pdf_path)
    questions = [pair['question'] for pair in qa_pairs]
    question_embeddings = embedding_model.encode(questions)

def find_similar_questions(user_question, question_embeddings, top_k=3, threshold=0.7):
    user_embedding = embedding_model.encode([user_question])
    similarities = cosine_similarity(user_embedding, question_embeddings)[0]
    sorted_indices = np.argsort(similarities)[::-1]
    return [(questions[i], similarities[i], i) for i in sorted_indices if similarities[i] >= threshold][:top_k]

def get_related_answer(user_question):
    similar_questions = find_similar_questions(user_question, question_embeddings)
    if similar_questions:
        context = " ".join(qa_pairs[i]['answer'] for _, _, i in similar_questions)
        result = qa_pipeline(question=user_question, context=context, handle_impossible_answer=True)
        return result['answer'] if result['score'] > 0.5 else "I'm not confident about the answer."
    return "Sorry, I couldn't find a relevant answer."

# Chatbot Interface
st.subheader("Ask a Question")
user_question = st.text_input("What would you like to know about the provided PDF?")

if user_question:
    answer = get_related_answer(user_question)
    st.write("Answer:", answer)

    # Language Translation Option
    with st.form("language_selection"):
        st.write("Select language for the answer:")
        dest_lang = st.selectbox("Select destination language:", ["en", "fr", "de", "es", "it", "ar"])
        submit_button = st.form_submit_button("Translate and Display Answer")

    # Handle translation
    if submit_button:
        try:
            translation = GoogleTranslator(source='auto', target=dest_lang).translate(answer)
            st.write(f"Translated Answer ({dest_lang}): {translation}")
        except Exception as e:
            st.error(f"An error occurred during translation: {str(e)}")
```
